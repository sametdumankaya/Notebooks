```python
df_geolocation = pd.read_csv('olist_geolocation_dataset.csv')
df_geolocation.head()
```

```python
df_order_items = pd.read_csv('olist_order_items_dataset.csv')
df_order_items.head()
```

```python
df_order_payments = pd.read_csv('olist_order_payments_dataset.csv')
df_order_payments.head()
```

```python
df_order_reviews = pd.read_csv('olist_order_reviews_dataset.csv')
df_order_reviews.head()
```

```python
df_orders = pd.read_csv('olist_orders_dataset.csv')
df_orders.head()
```

```python
df_products = pd.read_csv('olist_products_dataset.csv')
df_products.head()
```

```python
df_sellers = pd.read_csv('olist_sellers_dataset.csv')
df_sellers.head()
```

```python
df_prodct_category_translation = pd.read_csv('product_category_name_translation.csv')
df_prodct_category_translation.head()
```

### If there are any duplicate orders, we should have only one of them

```python
df_order_payments = df_order_payments.drop_duplicates('order_id', False)
```

### Merge different DataFrames into one big DataFrame

```python
df = pd.merge(df_orders, df_order_payments, on="order_id")
df = df.merge(df_customers, on="customer_id")
df = df.merge(df_order_items, on="order_id")
df = df.merge(df_products, on="product_id")
df = df.merge(df_prodct_category_translation, on="product_category_name")
df.head()
```

### Geolocation information is important in terms of delivery time. Farther customer means longer delivery time. Additionally, calculate the mean of the lat and lng values based on state

```python
df['customer_state'] = df['customer_state'].apply(lambda x : x.lower())
df['customer_city'] = df['customer_city'].apply(lambda x : x.lower())
geo_state = df_geolocation.groupby('geolocation_state')['geolocation_lat','geolocation_lng'].mean().reset_index()
geo_state['geolocation_state'] = geo_state['geolocation_state'].apply(lambda x : x.lower())
geo_city = df_geolocation.groupby('geolocation_city')['geolocation_lat','geolocation_lng'].mean().reset_index()
geo_city['geolocation_city'] = geo_city['geolocation_city'].apply(lambda x : x.lower())
geo_city.rename(columns={'geolocation_lat': 'mean_lat','geolocation_lng':'mean_lng'}, inplace=True)
df = pd.merge(df, geo_state, how='left', left_on='customer_state',right_on='geolocation_state')
df = pd.merge(df, geo_city,how='left',left_on='customer_city',right_on='geolocation_city')
df.head()
```

### Since only a very small piece of data is missing, we can drop the missing values safely

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True) 
```

### Change the date columns' data type to datetime so we can perform date and time operations 

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'])
df['order_delivered_carrier_date'] = pd.to_datetime(df['order_delivered_carrier_date'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])
```

### Calculate our target feature (shipment_time_in_days). Shipment time is equal to the time between order getting shipped and delivered.

```python
df['shipment_time_in_days'] =  (df['order_delivered_customer_date'] - df['order_delivered_carrier_date']).dt.days
df.head()
```

### Add a new feature: shipment_day_of_week_name, the hypothesis is that when an order is shipped on certain days (such as Friday which is close to weekend), shipment time might get longer.

```python
df['shipment_day_of_week_name'] = df['order_delivered_carrier_date'].dt.day_name()
df.head()
```

### Keep only a subset of features that we think might be useful in terms of prediction of the shipment time in days

```python
df = df[[
    'price',
    'freight_value',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm',
    'mean_lat',
    'mean_lng',
    'shipment_time_in_days',
    'shipment_day_of_week_name'
]]
df.head()
```

### Some of the features should be greater than zero. If not, remove them.

```python
df = df[(df['price'] > 0) & 
       (df['freight_value'] > 0) &
       (df['product_weight_g'] > 0) &
       (df['product_length_cm'] > 0) &
       (df['product_height_cm'] > 0) &
       (df['product_width_cm'] > 0)]
```

### We can engineer some new features by using existing ones

```python
df['order_sum'] = df['price'] + df['freight_value']
df['mean_product_price'] = df['order_sum'] / df['product_weight_g']
df['order_volume'] = df['product_length_cm'] * df['product_height_cm'] * df['product_width_cm']
```

### Let's assume that all of the orders are shipped from the same location and it is (0,0) coordinates. With this assumption, we can calculate the distance between shipment location and the customer location

```python
from geopy.distance import distance

df['assumed_distance'] = df.apply(lambda x: distance((x['mean_lat'], x['mean_lng']), (0, 0)).km, axis=1)
df.head()
```

```python
df['assumed_distance'].describe()
```

```python
sns.distplot(df['assumed_distance'])
```

```python
df = pd.get_dummies(df)
df.head()
```

### Plot the correlation heatmap

#### As seen in the heatmap, there are no strongly correlated features with delivery_time_in_days

```python
matplotlib.rcParams['figure.figsize'] = (15, 10)
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
```

### Seperate input and output columns and create train and test sets

```python
from sklearn.model_selection import train_test_split

X = df.loc[:, df.columns != 'shipment_time_in_days'].values
y = df.loc[:, df.columns == 'shipment_time_in_days'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
```

### Start modeling with LinearRegression

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

### Evaluate performance with 5-Fold Cross Validation using root mean squared error

```python
from sklearn.model_selection import cross_val_score

linear_regression_scores = cross_val_score(linear_regression, X_train, y_train, cv=5, scoring='neg_mean_squared_error') 
f'Average root mean squared error: {(-(sum(linear_regression_scores) / len(linear_regression_scores))) ** 0.5}'
```

### Now let's try popular Random Forest algorithm

```python
from sklearn.ensemble import RandomForestRegressor

random_forest = RandomForestRegressor()
random_forest.fit(X_train, y_train)
```

```python
random_forest_scores = cross_val_score(random_forest, X_train, y_train, cv=5, scoring='neg_mean_squared_error') 
f'Average root mean squared error: {(-(sum(random_forest_scores) / len(random_forest_scores))) ** 0.5}'
```

### Lastly try XGBoost on our data

```python
import xgboost

xgb_r = xgboost.XGBRegressor()
xgb_r.fit(X_train, y_train)
```

```python
xgb_scores = cross_val_score(xgb_r, X_train, y_train, cv=5, scoring='neg_mean_squared_error') 
f'Average root mean squared error: {(-(sum(xgb_scores) / len(xgb_scores))) ** 0.5}'
```

### The best performing model is RandomForestRegressor with the average root mean squared error of 7.33
